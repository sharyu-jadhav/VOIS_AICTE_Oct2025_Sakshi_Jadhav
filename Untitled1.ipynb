```python
# Calculate average price per neighborhood group
avg_prices = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)
print(avg_prices)
```

```python
# Convert to DataFrame for plotting
avg_price_df = avg_prices.reset_index()
avg_price_df.columns = ['Neighbourhood Group', 'Average Price']
```

```python
# Plot using color (no palette)
plt.figure(figsize=(10,6))
sns.barplot(data=avg_price_df, x='Neighbourhood Group', y='Average Price', color='skyblue')
plt.title("Average Price per Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.xlabel("Neighbourhood Group")
plt.show()
```

```python
# Check basic stats
df['Construction year'].describe()
```

```python
# Remove rows where Construction year is missing or zero
df_clean = df[df['Construction year'].notna() & (df['Construction year'] > 0)]
```

```python
# Create 5-year bins for construction year
bins = list(range(int(df_clean['Construction year'].min()), int(df_clean['Construction year'].max()) + 5, 5))
df_clean['year_bin'] = pd.cut(df_clean['Construction year'], bins)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_clean, x='Construction year', y='price', alpha=0.3, color='purple')
plt.title("Price vs Construction Year")
plt.xlabel("Construction Year")
plt.ylabel("Price ($)")
plt.show()
```

```python
correlation = df_clean[['Construction year', 'price']].corr()
print("Correlation between Construction Year and Price:\n", correlation)
```

```python
avg_price_year = df_clean.groupby('Construction year')['price'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=avg_price_year, x='Construction year', y='price', color='orange')
plt.title("Average Price by Construction Year")
plt.xlabel("Construction Year")
plt.ylabel("Average Price ($)")
plt.show()
```
