```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("1730285881-Airbnb_Open_Data.xlsx", engine='openpyxl')
```

```python
df.info()
df.isnull().sum()
```

```python
# Check all column names in the dataset
df.columns
```

```python
property_counts = df['room type'].value_counts()
print("Different property types and their counts:\n")
print(property_counts)
```

```python
# Count listings by neighborhood group
print("Listings by Neighborhood Group:\n")
print(df['neighbourhood group'].value_counts())
```

```python
# 1. What are the different property types in the dataset?
print("Property Types:\n", df['room type'].value_counts())
```

```python
# 2. Which neighborhood group has the highest number of listings?
print("Listings by Neighborhood Group:\n", df['neighbourhood group'].value_counts())
```

```python
# 3. Average price by neighborhood group
avg_price = df.groupby('neighbourhood group')['price'].mean()
print("Average Prices by Neighborhood Group:\n", avg_price)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='Construction year', y='price', data=df)
plt.title("Price vs Construction Year")
plt.xlabel("Construction Year")
plt.ylabel("Price")
plt.show()
```

```python
top_hosts = df.groupby('host name').size().sort_values(ascending=False).head(10)
print("Top 10 hosts by number of listings:\n", top_hosts)
```

```python
verified_reviews = df[df['host_identity_verified']=='Yes']['reviews per month'].mean()
unverified_reviews = df[df['host_identity_verified']=='No']['reviews per month'].mean()

print("Average Reviews per Month:")
print("Verified Hosts:", verified_reviews)
print("Unverified Hosts:", unverified_reviews)
```

```python
correlation = df[['price', 'service fee']].corr().iloc[0,1]
print("Correlation between Price and Service Fee:", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(x='room type', data=df)
plt.title("Property Types in Airbnb NYC")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

```python
avg_price_by_year = df.groupby('Construction year')['price'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Construction year', y='price', data=avg_price_by_year, marker='o')
plt.title("Average Price by Construction Year")
plt.xlabel("Construction Year")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate average price
avg_price = df.groupby('neighbourhood group')['price'].mean().reset_index()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='neighbourhood group', y='price', data=avg_price, palette='viridis')
plt.title("Average Airbnb Price by Neighborhood Group")
plt.xlabel("Neighborhood Group")
plt.ylabel("Average Price")
plt.show()
```

```python
# Create price bins
df['price_bin'] = pd.cut(df['price'], bins=10)

# Create pivot table
pivot = df.pivot_table(
    index='price_bin',
    columns='Construction year',
    values='id',
    aggfunc='count',
    fill_value=0,
    observed=False  # explicitly set to remove the warning
)

plt.figure(figsize=(15,6))
sns.heatmap(pivot, cmap="YlGnBu")
plt.title("Heatmap of Listings: Price vs Construction Year")
plt.xlabel("Construction Year")
plt.ylabel("Price Range")
plt.show()
```
